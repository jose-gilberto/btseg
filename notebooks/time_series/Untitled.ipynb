```python
from glob import glob
import pandas as pd
import btseg.transformers as t
import cv2
import matplotlib.pyplot as plt
```

```python
path_to_image_folders = glob('../../data/Outdir/*/')
path_to_image_folders[:5]
```

```python
path = '../../data/Outdir/BraTS20_Training_001/73.png'
```

```python
img = cv2.imread(path)
```

```python
transforms_region_1 = t.Compose([
    t.ConvertColor(conversion_code=cv2.COLOR_BGR2GRAY),
    t.RegionSelection(region=1),
])
```

```python
transformed_img = transforms_region_1(img)
plt.imshow(transformed_img)
```

```python
df1 = pd.read_csv('region_1_ts.csv')
ts1 = df1.iloc[0, :][1:]
ts1
```

```python
fig = plt.figure(figsize=(10,4))
ts1.plot()
```
