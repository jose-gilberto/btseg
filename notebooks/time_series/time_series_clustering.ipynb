````python
import pandas as pd
````

````python
time_series = pd.read_csv('./region_1_ts.csv')
time_series
````

````python
ts_ = time_series.dropna()
ts_ = ts_.drop(['id'], axis=1)
ts_
````

````python
X = []
for row in ts_.values:
    X.append(list(row))
````

## Hierarchical

````python
from dtaidistance import dtw
import numpy as np

X = np.array(X)
ds = dtw.distance_matrix_fast(X)
ds
````

```
from sktime.distances.elastic_cython import dtw_distance
import numpy as np

distance_matrix = np.zeros(shape=(len(X), len(X)))

for i in range(len(X)):
    for j in range(len(X)):
        x = np.array(X[i])
        y = np.array(X[j])
        if i != j:
            dist = dtw_distance(x, y)
            distance_matrix[i, j] = dist
            
distance_matrix
```

````python
import hdbscan

# hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=5,
#                                 metric='euclidean',
#                                 cluster_selection_method='leaf',
#                                 prediction_data=True)

hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=3,
                                cluster_selection_method='leaf',
                                metric='precomputed')
# hdbscan_model.fit(ts_)
hdbscan_model.fit(ds)

# hdbscan_model.fit(ts_)
````

````python
hdbscan_model.probabilities_
````

````python
out_labels = hdbscan_model.labels_
out_labels
````

````python
result_df = pd.DataFrame({
    'id': list(range(len(ts_))),
    'group': out_labels,
    'proba': hdbscan_model.probabilities_
})

result_df[result_df['group'] == 2].sort_values('proba', ascending=False).head()
````
