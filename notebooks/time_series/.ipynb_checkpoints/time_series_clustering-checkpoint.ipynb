```python
import pandas as pd
```

```python
time_series = pd.read_csv('./region_1_ts.csv')
time_series
```

```python
ts_ = time_series.dropna()
ts_ = ts_.drop(['id'], axis=1)
ts_
```

```python
X = []
for row in ts_.values:
    X.append(list(row))
X
```

```python
from tslearn.clustering import TimeSeriesKMeans

model = TimeSeriesKMeans(n_clusters=3, metric="dtw", max_iter=10)
model.fit(X)
```

```python
model.labels_
```

```python
from sklearn import cluster

spectral = cluster.SpectralClustering(
    n_clusters=3,
    eigen_solver="arpack",
    affinity="nearest_neighbors",
)
spectral.fit(X)
```

```python
spectral.labels_
```

```python
dbscan = cluster.DBSCAN(eps=0.3)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
birch = cluster.Birch(n_clusters=3)
birch.fit(X)
```

```python
birch.labels_
```
