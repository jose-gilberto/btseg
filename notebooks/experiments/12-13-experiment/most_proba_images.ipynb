```python
import pandas as pd
```

```python
result1 = pd.read_csv('./result_1.csv')
# Fixing some issues with 
result1['id'] = result1 + 1
result1['id'] = result1['id'].astype(int)

result1
```

```python
time_series_1 = pd.read_csv('./time_series_1.csv')
time_series_1
```

```python
results_for_plot = result1[result1.proba > 0.9].sort_values(['group'])
results_for_plot
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 4))
x_values = list(range(360))

for val in results_for_plot.query('group == 0').values:
    id, group, proba = val
    time_series = time_series_1.query(f'id == {id}').drop(['id', 'path', 'slice_path'], axis=1).values[0]
    plt.plot(x_values, time_series, color='darkslategray')
    

for val in results_for_plot.query('group == 1').values:
    id, group, proba = val
    time_series = time_series_1.query(f'id == {id}').drop(['id', 'path', 'slice_path'], axis=1).values[0]
    plt.plot(x_values, time_series, color='lightblue')

for val in results_for_plot.query('group == 2').values:
    id, group, proba = val
    time_series = time_series_1.query(f'id == {id}').drop(['id', 'path', 'slice_path'], axis=1).values[0]
    plt.plot(x_values, time_series, color='cadetblue')
    
plt.show()
```

```python
for g in [0, 1, 2]:
    for val in results_for_plot.query(f'group == {g}').values:
        id, group, proba = val
        print(group, '-', id, '-', proba)
        path = time_series_1.query(f'id == {id}')['slice_path'].values[0]
        print(path)
        print('='*10)
    print('\n\n')
```

```python
import btseg.transformers as t
import cv2

transforms = t.Compose([
    t.ConvertColor(conversion_code=cv2.COLOR_BGR2GRAY),
    t.RegionSelection(region=1),
    t.Resize(proportion_scale=True, g_measure=700),
    t.GaussianBlur(kernel_size=(7, 7), sigma_x=0),
    t.CannyEdge(threshold_1=50, threshold_2=100),
    t.Dilate(kernel=None, iterations=1),
    t.Erode(kernel=None, iterations=1),
    t.MinimumBoundingBox(),
])
```
