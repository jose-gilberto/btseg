```python
import pandas as pd
import numpy as np
from dtaidistance import dtw
from dtaidistance import ed
from btseg.clustering.hdbscan_clustering import HDBSCANClusterer
```

```python
df = pd.read_csv('time_series_1.csv')
df.head()
```

```python
X = df.drop(['id', 'path', 'slice_path'], axis=1).values
len(X)
```

```python
hdbscan = HDBSCANClusterer(min_cluster_samples=5, min_samples=5, metric='euclidean')
hdbscan.model
```

```python
hdbscan.fit(X)
```

```python
out_labels = hdbscan.model.labels_
out_labels
```

```python
counts = np.unique(out_labels, return_counts=True)
counts
```

```python
Xn = df.drop(['id', 'path', 'slice_path'], axis=1).copy()
Xn['label'] = out_labels
Xn = Xn.query('label != -1')
Xn['label'] = Xn['label'] + 1
Xn
```

```python
from DBCV import DBCV

score = DBCV(Xn.drop(['label'], axis=1).values, Xn['label'].values)
score
```
